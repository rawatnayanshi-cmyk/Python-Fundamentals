```python
#dictionary.
#dictionary are used to store data values in key:value pairs.
```

```python
dictionary = {"name": "A", "age": 20 , "Year": 2002}
print(dictionary)
```

```python
dict1 = {"brand":"BMW","Model":"ix","year":1964}
print(dict1)
```

```python
#accessing dict elements.
print(dict1["brand"])
```

```python
#for duplicate keys.
dict2 = {"brand":"BMW","Model":"ix","year":1964,"year":2024}
print(dict2)
```

```python
dict3 = {"class":2,"semester":1}
print(dict3)
```

```python
# for accessng key items.
x = dict2.get("year")
print(x)
```

```python
x = dict2.get("brand")
print(x)
```

```python
dict2.keys()
```

```python
dict2.values()
```

```python
#to change the values in dictionary.
dict1["year"]=2026
print(dict1)
```

```python
dict3["class"]=12
print(dict3)
```

```python
#to add new key value in dictionary.
dict2["color"]="white"
print(dict2)
```

```python
#the value in dictionary items can be of any datatype.
d1 = {"name":"Rishabh","year":2026,"Pass":True,"marks":[90,95,99]}
print(d1)
```

```python
d1.update({"year":2025})
print(d1)
```

```python
d1.update({"contactno":7685421558})
print(d1)
```

```python
#for removing dict items.
#the pop() method
d1.pop("contactno")
print(d1)
```

```python
#the popitem() method removes the last inserted item from the item.
d1.popitem()
print(d1)
```

```python
#the delete item removes the specified keys
del d1["year"]
print(d1)
```

```python
d1.clear()
print(d1)
```

```python
d2 = {"color":"blue","flower":"rose"}
print(d2)
```

```python
#COPY DICT
```

```python
d2 = {"color":"blue","flower":"rose"}
print(d2)
```

```python
d3 = d2.copy()
print(d3)
```

```python
d4=dict(d3)
print(d4)
```

```python
a = {"name":"Rishabh","year":2026,"Pass":True,"marks":[90,95,99]}
print(a)
```

```python
b = a.copy()
print(b)
```

```python
c = dict(b)
print(c)
```
