```python
#list item addition 
#1.append 2. insert 3. extend.
```

```python
course = ["SQL" , "Python" , "Stats"]
print(course)
```

```python
course.append("Excel")
print(course)
```

```python
subjects = ["English" , "hindi" , "maths"]
subjects.append("sst")
print(subjects)
```

```python
#insert item
#to insert a list item at specified index, use insert () method.
```

```python
course.insert(3,"Algo")
print(course)
```

```python
company = ["TCS" ,"Infosys" , "Wipro"] 
print(company)
```

```python
#extend list 
#to append elements from another list to the current list, use the extend().
```

```python
course.extend(company)
print(course)
```
