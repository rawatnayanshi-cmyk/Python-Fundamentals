```python
#array stacking and splitting
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
print(np.vstack((a,b)))
print(np.hstack((a,b)))
```

```python
c = np.array([[1,2,3],
            [4,5,6]])
print(c)
print(c.T)
```

```python
split = np.hsplit(c,3) 
for s in split:
    print(s)
```

```python
split = np.vsplit(c,2) 
for s in split:
    print(s)
```
