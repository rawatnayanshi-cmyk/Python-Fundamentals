Sorting matrices

```python
matrix_a = np.random.randint(10,size = (4,4))
matrix_a
```

```python
matrix_a_sort_default = np.sort(matrix_a)
matrix_a_sort_default
```

```python
# same as default
matrix_a_sort_axis = np.sort(matrix_a, axis =  -1)
matrix_a_sort_axis
```

```python
# sort "along" the rows, meaning hold the column fixed and sort the values from the different rows (i.e., sort within each column)
matrix_a_sort_axis = np.sort(matrix_a, axis = 0)
matrix_a_sort_axis
```

Sometimes, we are missing data! This happens a lot when we're collecting data. It's important to represent this clearly in our data. One way to do this is with a "NaN"- short for "not a number". numpy has a special way of representing this. Let's work with an example.

```python
arr = np.array([[1, 2, np.nan],
                [4, np.nan, 6],
                [7, 8, 9]])
print(arr)
```

```python
np.isnan(arr)
```

Counting NaN values

```python
np.count_nonzero(np.isnan(arr))
```

```python
np.isnan(arr).sum()
```

Working with NaN values

```python
arr.sum()
```

```python
np.nansum(arr)
```

```python
arr[np.isnan(arr)] = 0
```

```python
np.where(np.isnan(arr), 0, arr)
```
