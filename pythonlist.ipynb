```python
#extend list 
#to append elements from another list to the current list, use the extend().
```

```python
course = ["SQL" , "Python" , "Stats", "Excel"]
company = ["TCS" ,"Infosys" , "Wipro"] 
course.extend(company)
print(course)
```

```python
fruits = ["apple" , "blueberries" , "muskmelon"]
countries = ["Spain","India","America"]
fruits.extend(countries)
print(fruits)
```

```python
colors = ["pink", "blue","black"]
subjects = ["English" , "hindi" , "maths"]
colors.extend(subjects)
print(colors)
```

```python
applications = ["Instagram" , "Facebook" , "Google"]
software = ["DBMS" , "SQL" , "MongoDB"] 
applications.extend(software)
print(applications)
```

```python
name1 = ["Sakshi" , "Riya" , "Neha"]
name2 = ["Rahul" , "Krish" , "Himanshu"]
name1.extend(name2)
print(name1)
```
