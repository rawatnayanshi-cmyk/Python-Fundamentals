```python
#union method
set1 = {"x" ,"y","z"}
set2 = {1,2,3,}
set3 = set1.union(set2)
print(set3)
```

```python
set1 = {"x" ,"y","z"}
set2 = {1,1,2,3,3}
set3 = set1.union(set2)
print(set3)
```

```python
# | operator
set1 = {"x" ,"y","z","a"}
set2 = {1,2,3,}
set3 = set1|(set2)
print(set3)
```

```python
#join multiple sets
set1 = {"x" ,"y","z","a"}
set2 = {1,2,3,}
set3 = {True,False}
set4= {"king","kong"}
largeset = set1.union(set2,set3,set4)
print(largeset)
```

```python
#intersection method.
set1 = {"x" ,"y","z","a"}
set2 = {1,2,3,"z"}
set3 = set1.intersection(set2)
print(set3)
```

```python
# & operator
set1 = {"x" ,"y","z","a"}
set2 = {1,2,3,"z"}
set3 = set1 & set2
print(set3)
```

```python
#difference method.
set1 = {"x" ,"y","z","a"}
set2 = {1,2,3,"z","a"}
set3 = set1.difference(set2)
print(set3)
```
